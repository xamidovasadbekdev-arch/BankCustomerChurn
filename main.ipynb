```python
df.sample(3)
```

```python
sns.barplot(x="Geography", y="Exited", data=df)

plt.xticks(rotation=90)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))

sns.regplot(x="Age", y="Exited", data=df, ax=ax1, color="green")
ax1.set_title("Age vs Exited in Regplot")

sns.violinplot(x="NumOfProducts", y="Exited", data=df, ax=ax2)
ax2.set_title("NumOfProducts vs Exited in Violinplot")

plt.tight_layout()
plt.show()
```

```python
df.groupby("NumOfProducts")["Exited"].agg(["mean", "count"])
```
